```python
import matplotlib as mpl
import matplotlib.pyplot as plt
# developing distribution by gender
gender_distribution = df['Gender'].value_counts()
print(gender_distribution)

```

```python
counting_genders = df['Gender'].value_counts()

plt.figure(figsize=(5, 4))
plt.pie(
    counting_genders,
    labels=counting_genders.index,
    autopct='%1.1f%%',
    startangle=40,
    colors=['blue', 'red']
)

plt.title('Gender Distribution', fontsize=10)
plt.show()
```

```python
frequencies_of_purchases = df['Frequency of Purchases'].value_counts()
print(frequencies_of_purchases)
```

```python
purchases_frequency = df['Frequency of Purchases'].value_counts()
frequency = purchases_frequency.index
quantity = purchases_frequency.values

fig, ax = plt.subplots(figsize=(8, 6))

bar_colors = ['blue', 'yellow', 'navy', 'red', 'orange', 'cyan', 'lime']
ax.bar(frequency, quantity, color=bar_colors)

ax.set_xlabel('Frequency of Purchases')
ax.set_ylabel('Quantity')
ax.set_title('Purchases Frequency')

plt.show()
```

```python
location = df['Location'].value_counts()
print(location)
```

```python
regions = {
    'North': ['Illinois', 'Indiana', 'Iowa', 'Michigan', 'Minnesota', 'Nebraska', 'North Dakota', 'Ohio', 'South Dakota', 'Wisconsin'],
    'West': ['Alaska', 'Arizona', 'California', 'Colorado', 'Hawaii', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Oregon', 'Utah', 'Washington', 'Wyoming'],
    'South': ['Alabama', 'Arkansas', 'Delaware', 'Florida', 'Georgia', 'Kentucky', 'Louisiana', 'Maryland', 'Mississippi', 'Kansas', 'Missouri', 'North Carolina', 'Oklahoma', 'South Carolina', 'Tennessee', 'Texas', 'Virginia', 'West Virginia'],
    'East': ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'New Jersey', 'New York', 'Pennsylvania', 'Rhode Island', 'Vermont']
}

def get_region(state):
  for region, states in regions.items():
    if state in states:
      return region
  return None

df['Region'] = df['Location'].apply(get_region)

region_counts = df['Region'].value_counts()
print(region_counts)
```

```python
region_counts = df['Region'].value_counts()

plt.figure(figsize=(5, 4))
plt.pie(
    region_counts,
    labels=region_counts.index,
    autopct='%1.1f%%',
    startangle=40,
    colors=['sienna', 'salmon', 'cyan', 'khaki']
)

plt.title('Region', fontsize=10)
plt.show()
```
